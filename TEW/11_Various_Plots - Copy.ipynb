## 11 Histogram, Box plots and Scatter plots using Seaborn and Matplotlib in Python

Statistical plots always allow viewers to identify outliers(?),  visualised distributions, deduce variable types and discovers relationships & correlations between variables in a dataset.<br>
We are going to look at how to use statistical plots to detect outliers, reduced variable distribution & tie(?) and uncover relationships and correlations.

- Histograms
    - Shows a variable distribution as a set of adjacent rectangles on a data chart
    - Represents counts of data within a numerical range of values
- Scatter plots
    - Are useful for exploring the interrelations or dependencies between two different variables
    - Ideal for visually spotting outliers and trends in data
- Box plots
    - Show variable spread 
    - Useful for outlier detection

```python
#As always, we first set up our JN with our imports and settings...
import numpy as np
from numpy.random import randn

import pandas as pd
from pandas import Series, DataFrame
from pandas.tools.plotting import scatter_matrix # This allows us to make scatter plots

#For data visualisation
import matplotlib.pyplot as plt
from pylab import rcParams 
import seaborn as sb

#To display matplotlib plots in JN
%matplotlib inline

# This allows you to make changes to the stylesheet that underlies matplotlib 
rcParams['figure.figsize'] = 10, 8

#seaborn settings
sb.set_style('whitegrid')
```

### Importing our data to plot

```python
address = 'D:/Python-For-Data-Science-Training-TEW/datasets/mtcars.csv'
```

```python
cars = pd.read_csv(address, sep = '\t')

cars.index = cars['model']

mpg = cars.mpg
```

```python
cars.head()
```

### Creating a Histogram

```python
mpg.plot(kind = 'hist')
```

```python
plt.hist(mpg)

plt.plot() # If we don't do this then the histogram plots but you get the underlying array printed out too
```

#### Using Seaborn
We are going to plot the same data only this time, we are going to use the seaborn plotting function. Seaborn is great for statistical plotting.

To plot a series or variable in seaborn, you call the ``distplot`` function in seaborn and pass it the variable that you want to plot...

```python
sb.distplot(mpg)
```

This gives us a trend line and shows that the distribution as a line. This is done for us automatically.

## Scatterplot in Matplotlib

```python
cars.plot(kind = 'scatter', x = 'hp', y = 'mpg', c = ['darkgray'], s = 150 )
```

### To do the same in Seaborn

```python
sb.regplot(x = 'hp', y = 'mpg', data = cars, scatter = True)
```

``data = cars`` - tells seaborn which dataset we want to use
Once again, seaborn automatically creates a trend line

### Generatinbg a scatter plot matrix

```python
sb.pairplot(cars)
```

### Categorical Colouring 
Adding a third dimension of information to this chart

```python
cars_df = pd.DataFrame((cars.iloc[:, (1,3,4,6)].values), columns = ['mpg', 'disp', 'hp', 'wt'])

cars_target = cars.iloc[:, 9] # Selecting column 9
target_names = [0, 1] # zero for automatic transmission & 1 for manual

#Third dimension of information - our ctaegorical variable
cars_df['group'] = pd.Series(cars_target, dtype = 'category') # We create a new Series/Column called group using the cars_target
sb.pairplot(cars_df, hue = 'group', palette = 'hls')
```
