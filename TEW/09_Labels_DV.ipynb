## 9. Labels and Annotations using Matplotlib and Seaborn in Python
Labels and annotation add a deeper layer of context to a plot enabling the plot to convey extra meaning to its viewers. Here we are going to look at...
- How to label plot features
- Adding a legend to a plot 
- Annotating your plot
There are two methods of labelling and annotating your plots - Functional and OOP 

### Annotataing your plot
**To annotate some features of the plot
``.annotate(xy, xytext and arrowprop)``
- **Parameters**
    - xy - The location that you want to annotate
    - xytext - The location of where we want the text edit 
    - arrowprop - This defines the particulars of the arrow that we use


**Adding a Legend**<br>
Placing a legend on plot axes
- .legend(label & location )

**Parameters**
- Label - variable or category labels
- loc - the location of the legend<br>
For where you want that legend to be added

```python
#As always, we set up our JN with our imports and settings...
import numpy as np
from numpy.random import randn

import pandas as pd
from pandas import Series, DataFrame

#For data visualisation
import matplotlib.pyplot as plt
from pylab import rcParams 
import seaborn as sb

#To display matplotlib plots in JN
%matplotlib inline

# This allows you to make changes to the stylesheet that underlies matplotlib 
rcParams['figure.figsize'] = 5, 4

#seaborn settings
sb.set_style('whitegrid')
```

### Labelling plot features
**The functional method**<br>

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

# To create a matplotlib bar chart using x & y
plt.bar(x, y)

# Labelling the x axis
plt.xlabel('Your X Axis Label') # Simply pass a string to the matplotlib xlabel function

# Lableiing the y axis
plt.ylabel('Your Y Axis Label')

# When we print this out...
```

We can see that the labels have printed out perfectly

**Labelling a Piechart**<br>
This is slightly different from the above method...

```python
# We add a new object and enter the values that we want plotted...
z = [1,2,3,4,5]

# We are going to create some labels for each element in the pie chart
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

# Plotting our pie chart
plt.pie(z, labels = veh_type)
plt.show()
```

**The Object Orientated Method**<br>
To demo the OOP method, we are goiing to use the mtcars dataset once again

```python
#Loading in our datset...loving setting the address as a variable
address = 'D:/Python-For-Data-Science-Training-TEW/datasets/mtcars.csv'

cars = pd.read_csv(address, sep = '\t')

# Creat a variable that holds the mpg column data
mpg = cars.mpg
cars.head()
```

**Now we go through the OOP specific methods...**

```python
fig = plt.figure()

# We use the add_axes function to add the axis and pass in the details of where we want the axis to be added
ax = fig.add_axes([.1, .1, 1, 1])

# Plotting mpg variable
mpg.plot()

# Setting x tick marks going from 1-31 
ax.set_xticks(range(32))

# Adding custom string labels - We are going to use the model column from our df but you could pass in a string
ax.set_xticklabels(cars.model, rotation = 60, fontsize = 'medium')

#Adding a title to the chart
ax.set_title('Miles per Gallon for the cars in MTCARS dataset')

# Labelling the X & Y axis
ax.set_xlabel('Car Names')
ax.set_ylabel('Miles/Gallon')
```

OK this took a few goes to get right but you will notice that our plot...
- Has a title
- Both the X and Y axes have a label
- Along the X axis, every car name is shown
    - This does look a bit messy tho
    - We can clearly see what the rotation parameter does to the car name

### Adding a legend to your plot
**The Functional Method**

```python
plt.pie(z)

# We call the legend function on the list that defines the legend (?)
plt.legend(veh_type, loc = 'best')
# To position the legend, you must pass in a location argument. Options include best, upper right, upper left, etc

plt.show()
```

**The Object Orientated Method**<br>
We will re-use the mpg variable that we created from the mpg column in the cars df.

**Now we go through the OOP specific methods...**

```python
fig = plt.figure()

# We use the add_axes function to add the axis and pass in the details of where we want the axis to be added
ax = fig.add_axes([.1, .1, 1, 1])

# Plotting mpg variable
mpg.plot()

# Setting x tick marks going from 1-31 
ax.set_xticks(range(32))

# Adding custom string labels - We are going to use the model column from our df but you could pass in a string
ax.set_xticklabels(cars.model, rotation = 60, fontsize = 'medium')

#Adding a title to the chart
ax.set_title('Miles per Gallon for the cars in MTCARS dataset')

# Labelling the X & Y axis
ax.set_xlabel('Car Names')
ax.set_ylabel('Miles/Gallon')

# Adding and positioning the legend
ax.legend(loc = 'best')
```

We can see that the legend has been added to the upper right hand cormer as the best position for it to be. This is a judgement that Python makes when you use the best argument. It could easily have gone for the upper left as there are no plottings to get in the way there.

## Annotating your plot
We are going to add some annotation to call up the max value of the mgp variable

```python
# To find the max value of the mpg variable...
mpg.max()
```

```python
fig = plt.figure()

# We use the add_axes function to add the axis and pass in the details of where we want the axis to be added
ax = fig.add_axes([.1, .1, 1, 1])

# Plotting mpg variable
mpg.plot()

#Setting the title
ax.set_title('Miles per Gallon for the cars in MTCARS dataset')

# Setting the Y label
ax.set_ylabel('Miles/Gallon')

ax.set_ylim([0, 45])

# Now we are going to annotate the plot
ax.annotate('Toyota Corolla', xy = (19, 33.9), xytext = (21, 35), arrowprops = dict(facecolor = 'black', shrink = 0.05))
# Toyota Corolla - Is the string for the actual annotation
# xy - Is the location where the annotation will sit
# xytext - Specify the location where you want the annotation text to be placed 21 being the x figure and 35 for y
# arrowprops - This specifies the properties of the arrow that we are drawing from the text to the point we are annotating  
```
